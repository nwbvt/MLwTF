```python
import tensorflow as tf
import numpy as np
```

```python
raw_data = np.random.normal(10, 1, 100)
alpha = tf.constant(.05)
curr_value = tf.placeholder(tf.float32)
prev_avg = tf.Variable(0.)
update_avg = alpha * curr_value + (1-alpha) * prev_avg

avg_history = tf.summary.scalar("running_avg", update_avg)
value_history = tf.summary.scalar("incoming_values", curr_value)
merged = tf.summary.merge_all()
writer = tf.summary.FileWriter('./logs')

init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for datum,i in zip(raw_data, range(len(raw_data))):
        summary_str, curr_avg = sess.run([merged, update_avg], feed_dict={curr_value: datum})
        sess.run(tf.assign(prev_avg, curr_avg))
        print(datum, curr_avg)
        writer.add_summary(summary_str, i)
```
